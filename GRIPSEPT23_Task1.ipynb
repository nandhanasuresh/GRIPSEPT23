### AIM : To predict the percentage score of a student based on the number of study hours.

```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# Data source URL

data_url = "http://bit.ly/w-data"
```

```python
# Load the dataset into a Pandas DataFrame

df = pd.read_csv(data_url)
```

```python
# Display the first four entries in the dataset

df.head(4)
```

```python
# Display the last four entries in the dataset

df.tail(4)
```

```python
# Get information about the dataset

df.info()
```

```python
# Descriptive statistics of the dataset

df.describe()
```

```python
# Shape of the dataset (number of rows and columns)

df.shape
```

```python
# Check for null values in the dataset

df.isnull().sum()
```

##### There are no null values in the dataset.


```python
# Plotting the dataset

plt.style.use('ggplot')
df.plot(kind='line')
plt.title("Hours vs Percentage Score", color='blue')
plt.xlabel('Hours Studied per day')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# Scatter plot to visualize the distribution of scores

plt.scatter(x=df.Hours, y=df.Scores)
plt.title('Hours vs Percentage Score', color='blue')
plt.xlabel('Hours Studied per day')
plt.ylabel('Percentage Score')
plt.show()
```

##### From the graphs, we can see a positive relationship between the percentage scores and hours of study.


```python
# Data preparation

x = df.iloc[:, :-1].values  # Independent variable (Hours)
y = df.iloc[:, 1].values    # Dependent variable (Scores)
```

```python
# Split the dataset into training and testing sets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Model training

model = LinearRegression()
model.fit(x_train, y_train)
print("Training completed")
```

```python
# Plotting the regression line

line = model.coef_ * x + model.intercept_
```

```python
# Plotting for test data

plt.scatter(x, y)
plt.plot(x, line, color="purple")
plt.show()
```

```python
# Making predictions

y_pred = model.predict(x_test)
```

```python
# Display actual vs. predicted values

df = pd.DataFrame({'Actual value': y_test, "Predicted value": y_pred})
df
```

```python
# Predicting the score for a student who studies for 9.25 hours per day

given_hours = 9.25
predicted = model.predict([[given_hours]])
print("Number of hours =", given_hours)
print("Predicted score is =", predicted[0])
```

```python
# Evaluating the model

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

##### MAE is a measure of how accurate the model is in predicting scores.
